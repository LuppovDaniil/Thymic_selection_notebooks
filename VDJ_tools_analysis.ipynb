```python
#VDJtools output for KECK
res_keck = pd.read_csv('~/vdj_tools_data/OLGA_1/KECK/keck_vdj_nt_single.txt', sep = '\t')
res_keck
```

```python
# adj p additing
from statsmodels.stats.multitest import multipletests

res_keck.insert(19, 'p.value.g_adj' ,multipletests(res_keck['p.value.g'])[1])
```

```python
fold_keck = ((res_keck['degree.s'] + 1)/res_keck['group.count.s'])/((res_keck['degree.c'] + 1)/res_keck['group.count.c'])
log2_fc_keck = np.log2(fold_keck)
log10_p_keck = -np.log10(res_keck['p.value.g'].apply(lambda x: x if x > 10**(-16) else 10**(-16)))
plt.scatter(log2_fc_keck, log10_p_keck, s = 1)
plt.xlabel('log2_fc')
plt.ylabel('log10_p')
plt.title('KECK vs OLGA ')
```

```python
# paint clusters with glycosylation sutes in red
import re
c_keck = res_keck['cdr3aa'].apply(lambda x: 'r' if bool(re.search(r'N\w[T, S]', x)) else 'grey')
alpha_keck = res_keck['cdr3aa'].apply(lambda x: 1 if bool(re.search(r'N\w[T, S]', x)) else 0.05)
s_keck = res_keck['cdr3aa'].apply(lambda x: 5 if bool(re.search(r'N\w[T, S]', x)) else 1)

c_keck
```

```python
# clusters representation

import pandas as pd
import numpy as np
import igraph as ig
from scipy.spatial.distance import pdist, squareform
from plotnine import *

def seqs2hamming(seqs, threshold = 1):
    # pairwise hamming
    def hdist(s1, s2):
        if len(s1) != len(s2):
            return float('inf')
        else:
            return sum(c1 != c2 for c1, c2 in zip(s1, s2))
    
    seqs = np.array(seqs).astype("str")
    dm = squareform(pdist(seqs.reshape(-1, 1), metric = lambda x, y: hdist(x[0], y[0])))
    dmf = pd.DataFrame(dm, index=seqs, columns=seqs).stack().reset_index()
    dmf.columns = ['id1','id2','distance']
    dmf = dmf[dmf['distance'] <= threshold]
    
    # graph
    graph = ig.Graph.TupleList(dmf[['id1', 'id2']].itertuples(index=False))
    
    # clusters
    clusters = graph.components()
    membership = clusters.membership
    
    # layout
    #layout = graph.layout('graphopt')
    layout = graph.layout('DrL')
    coords = np.array(layout.coords)
    
    df_graph = pd.DataFrame(
        {'cdr3': graph.vs()['name'],
         'cluster': membership,
         'x': coords[:,0],
         'y': coords[:,1]
    })
    
    # summary
    df_graph_summary = df_graph.groupby(['cluster']).agg(
        cluster_size = ('cluster', 'size'), 
        x_mean = ('x', 'mean'), 
        y_mean = ('y', 'mean')).reset_index()
    
    return pd.merge(df_graph, df_graph_summary)
```

```python
# top 10000 seqs to plot on clustermap
olga_dif_seqs = res_olga.sort_values(by = 'p.value.g').head(10000)['cdr3aa']
res = seqs2hamming(olga_dif_seqs)
```

```python
#clusters of enriached seqs in OLGA

res['cc'] = res['cluster'].astype("str")

g = (ggplot(res[res['cluster_size'] > 1], aes(x='x', y='y', color='cc')) + 
     geom_point(data = res[res['cluster_size'] == 1], color = 'grey') +
     geom_point() + 
     scale_color_discrete(guide=False) +
     xlab('') +
     ylab('') +
     theme_void() +
     theme(aspect_ratio = 1))
print(g)
```

```python
#res['cluster'].value_counts().head(5)
dif_clusters = res['cluster'].value_counts().head(5).index

import logomaker

for cluster in dif_clusters:
    clones = res[res['cluster'] == cluster].cdr3
    mat_df = logomaker.alignment_to_matrix(clones)
    logomaker.Logo(mat_df, color_scheme='skylign_protein', ax=None)
```
